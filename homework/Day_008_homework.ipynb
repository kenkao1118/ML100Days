```python
print('AMT_INCOME_TOTAL MEAN',app_train['AMT_INCOME_TOTAL'].mean())
print('AMT_INCOME_TOTAL STD',app_train['AMT_INCOME_TOTAL'].std())
```

```python
print('AMT_CREDIT mean',app_train['AMT_CREDIT'].mean())
print('AMT_CREDIT STD',app_train['AMT_CREDIT'].std())
```

```python
app_train['AMT_ANNUITY'].describe()
```

```python
a = app_train['AMT_CREDIT']
bins = np.arange(0,1000000,50000)
plt.hist(a, bins, facecolor='blue',edgecolor='black',alpha=0.7) 
plt.ylabel('Dollars')
plt.title("AMT CREDIT") 
plt.show()
```

```python
a = app_train['AMT_ANNUITY']
bins = np.arange(0,100000,5000)
plt.hist(a, bins, facecolor='green',edgecolor='black',alpha=0.7) 
plt.ylabel('Dollars')
plt.title("AMT ANNUITY") 
plt.show()
```
